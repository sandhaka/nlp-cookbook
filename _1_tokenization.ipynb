### Tokenization
Transform a sentence into a sequence of tokens (words).

Before start, test pytorch installation:

```python
import torch
x = torch.rand(5, 3)
print(x)
```

#### BERT

```python
from pytorch_transformers import BertTokenizer
```

```python
tokenizer = BertTokenizer.from_pretrained('bert-base-cased')
example_text = 'I\'ll go to bed late tonight'
tokenized = tokenizer.tokenize(example_text)
print(tokenized)
```

Using transformers package

```python
from transformers import BertTokenizerFast
tokenizer = BertTokenizerFast.from_pretrained('bert-base-cased')
tokenized = tokenizer('I\'ll go to bed late tonight', padding='max_length', max_length=512, truncation=True)
print(tokenized)
```
